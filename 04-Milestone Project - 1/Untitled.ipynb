```python
import random
class TicTacToe:
    def __init__(self):
        self.board = self.make_board()
        #self.marker = marker
        #self.position = position
        
        #self.winner = None
    @staticmethod
    def make_board():
        return [' ' for _ in range(1,11)]
    
    def display_board(self):
        #clear_output()  # Remember, this only works in jupyter!
        print('    |    |')
        print(' ' + self.board[7] + '  | ' + self.board[8] + '  | ' + self.board[9])
        print('    |    |')
        print('-------------')
        print('    |    |')
        print(' ' + self.board[4] + '  | ' + self.board[5] + '  | ' + self.board[6])
        print('    |    |')
        print('-------------')
        print('    |    |')
        print(' ' + self.board[1] + '  | ' + self.board[2] + '  | ' + self.board[3])
        print('    |    |')
    
    def player_input(self):
        marker = ' '
        while not (marker == 'X' or marker == 'O'):
            marker = input('Player 1: Do you want to be X or O? ').upper()
            
            if marker == 'X':
                return ('X', 'O')
            else:
                return ('O', 'X')
            
    def place_marker(self, marker, position):
        self.board[position] = marker
        
    def win_check(self, marker):
    
        return ((self.board[7] == marker and self.board[8] ==  marker and self.board[9] == marker) or # across the top
        (self.board[4] ==  marker and self.board[5] == marker and self.board[6] == marker) or # across the middle
        (self.board[1] ==  marker and self.board[2] == marker and self.board[3] == marker) or # across the bottom
        (self.board[7] ==  marker and self.board[4] == marker and self.board[1] == marker) or # down the middle
        (self.board[8] ==  marker and self.board[5] == marker and self.board[2] == marker) or # down the middle
        (self.board[9] ==  marker and self.board[6] == marker and self.board[3] == marker) or # down the right side
        (self.board[7] ==  marker and self.board[5] == marker and self.board[3] == marker) or # diagonal
        (self.board[9] ==  marker and self.board[5] == marker and self.board[1] == marker)) # diagonal
        
    def choose_first(self):
        if random.randint(0, 1) == 0:
            return 'Player 2'
        else:
            return 'Player 1'
    
    #def space_check(self):    
    #    return self.board[self.position] == ' '
    
    def full_board_check(self):
        for i in range(1,10):
            if self.board[i] == ' ':
                return False
        return True

    def player_choice(self):
        position = 0
    
        while position not in [1,2,3,4,5,6,7,8,9] or not  self.board[position] == ' ':
            position = int(input('Choose your next position: (1-9) '))
        return position
    
    def replay(self):
        return input('Do you want to play again? Enter Yes or No: ').lower().startswith('y')
```

```python
g = TicTacToe()
#g.display_board()
g.player_choice()
```

```python
def play(g):
    print('Welcome to Tic Tac Toe!')

    while True:
        # Reset the board
        # theBoard = [' '] * 10
        player1_marker, player2_marker = g.player_input()
        turn = g.choose_first()
        print(turn + ' will go first.')
    
        play_game = input('Are you ready to play? Enter Yes or No.')
    
        if play_game.lower()[0] == 'y':
            game_on = True
        else:
            game_on = False

        while game_on:
            if turn == 'Player 1':
                # Player1's turn.
            
                g.display_board()
                position = g.player_choice()
                g.place_marker(player1_marker, position)

                if g.win_check(player1_marker):
                    g.display_board()
                    print('Congratulations! You have won the game!')
                    game_on = False
                else:
                    if g.full_board_check():
                        g.display_board()
                        print('The game is a draw!')
                        break
                    else:
                        turn = 'Player 2'

            else:
                # Player2's turn.
            
                g.display_board()
                position = g.player_choice()
                g.place_marker(player2_marker, position)

                if g.win_check(player2_marker):
                    g.display_board()
                    print('Player 2 has won!')
                    game_on = False
                else:
                    if g.full_board_check():
                        g.display_board()
                        print('The game is a draw!')
                        break
                    else:
                        turn = 'Player 1'

        if not g.replay():
            break
    if __name__ == '__main__':
    #o_player = HumanPlayer('O')
    #x_player = RandonComputerPlayer('X')
        t = TicTacToe()
        play(t)
```
